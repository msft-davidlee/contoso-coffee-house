```
import csv
import json
import time
# Class to hold the individual record
class record:
    def __init__(self, memberId, transactionDate, transactionId):
        self.memberId = memberId
        self.transactionDate = transactionDate
        self.transactionId = transactionId
        # List (array) to hold the line items
        self.lineitems = []
# Class to represent the line item
class lineitem:
    def __init__(self, sku, roundedAmountSpent):
        self.sku = sku
        self.roundedAmountSpent = roundedAmountSpent
def csv_to_json(csvFilePath, jsonFilePath):
    jsonArray = []
    #read csv file
    with open(csvFilePath, encoding='utf-8') as csvf: 
        #load csv file data using csv library's dictionary reader
        csvReader = csv.DictReader(csvf) 
        # Created first record with garbage data (but needed for accessibility)
        holdingrecord = record("hold", "hold", "hold")
        # In place to prevent putting the garbage data in our array
        firstline = True
        #convert each csv row into python dict
        for row in csvReader: 
            # If the transaction id we just read matches our current transaction, just add the line item to the transactions lineitems list
            if row["transactionId"] == holdingrecord.transactionId:
                currentlineitem = lineitem(sku=row['sku'], roundedAmountSpent=row['roundedAmountSpent'])
                holdingrecord.lineitems.append(currentlineitem)
            else:
            # Otherwise add the record to the main list (unless it's the first row)
            # Create a new record based on the row, and add the line item from the row to the record
                if not firstline:
                    jsonArray.append(holdingrecord)
                holdingrecord = record(memberId=row['memberId'], transactionDate=row['transactionDate'], transactionId=row["transactionId"])
                currentlineitem = lineitem(sku=row['sku'], roundedAmountSpent=row['roundedAmountSpent'])
                holdingrecord.lineitems.append(currentlineitem)
                firstline = False
    #convert python jsonArray to JSON String and write to file
    with open(jsonFilePath, 'w', encoding='utf-8') as jsonf: 
        jsonString = json.dumps(jsonArray, indent=4, default=vars)
        jsonf.write(jsonString)

#replace csvFilepath with path of file generated in notebook #1
#replace jsonFilePath with name of new JSON file to be generated
csvFilePath = '/dbfs/mnt/<filename of CSV generated in Notebook #1>'
jsonFilePath = '/dbfs/mnt/<choose filename ending in .json to write into Blob Storage>'
start = time.perf_counter()
csv_to_json(csvFilePath, jsonFilePath)
finish = time.perf_counter()
```
